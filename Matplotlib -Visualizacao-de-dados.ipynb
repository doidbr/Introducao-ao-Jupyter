### Como importar matplotlib, pylab, pyplot, numpy e tudo mais  <a name="Imp-Matplotlib"></a>

O submódulo `matplotlib.pyplot` fornece uma interface orientada a objetos para a biblioteca de construção de gráficos. Muitos dos exemplos na documentação da matplotlib seguem a convenção de importação para importar `matplotlib.pyplot` como `plt` e `numpy` como `np`. É claro que é decisão inteiramente do usuário se ele vai importar a biblioteca `numpy` sob o nome `np` (como muitas vezes feito em exemplos matplotlib) ou como outro nome qualque a gosto do usuário. Nesse caso é recomendado que o usuário crie um padrão e mantenha ele, assim evitará problemas de portabilidade do seus códigos no futuro. Da mesma forma, é uma questão de gosto se o submodule gráficos (`matplotlib.pyplot`) são importados como `plt` como é feito na documentação matplotlib ou `plot` (o que poderia ser argumentado é um pouco mais claro), etc.


Como sempre um equilíbrio tem que ser encontrado entre preferências pessoais e consistência com prática comum em escolher estes nome. A coerência com o uso comum é naturalmente mais importante se o código é susceptível de ser utilizado por outros ou se ele for publicado.

Fazer um gráfico quase sempre necessita de matrizes de dados numéricos e é por isso que o módulo `numpy` é usado muito: ele fornece um gerenciamento de matriz muito rápido e eficiente para Python (ver [capítulo 15](15-numpy.ipynb)).

Poderíamos também ter escrito o exemplo 1a acima como no exemplo 1b (que é idêntico em funcionalidade ao exemplo acima e criará o mesmo gráfico):

````python
# Exemplo 1b
import pylab
import numpy as np  # Usando um novo nome para a NumPy

x = np.arange (-3.14, 3.14, 0.01)
y = np.sin(x)

pylab.plot(x, y)
pylab.plot(x, np.cos(x))
````

Como os objetos `numpy.arange` e `numpy.sin` já foram importados para o espaço de nomes (conveniência) `pylab`, também podemos escrevê-lo como exemplo 1c:

````python
# Exemplo 1c
import pylab as plt

x = plt.arange(-3.14, 3.14, 0.01)
y = plt.sin(x)
xmin = -3.0
xmax = +3.0
ymin = -1.0
ymax = +1.0
v = [xmin, xmax, ymin, ymax] # lista contendo os intervalos do gráficos
plt.axis(v)                  # lista contendo os intevalos dos eixos 
#plt.grid(b='on',axis='both')
plt.grid()
plt.plot(x, y)
plt.show()
````

Isso pode ser extremamente conveniente, mas vem com um grande aviso de perigo:

- Ao usar `from pylab import *` é aceitável no prompt de comando para criar interativamente gráficos e analisar dados, isso nunca deve ser usado em qualquer script para fazer gráficos.

- O nível mais alto do pylab fornece mais de 800 objetos diferentes que são todos importados para o espaço de nome global ao executar `from pylab import *`. Isso não é uma boa prática e pode entrar em conflito com outros objetos que já existem ou que serão criados posteriormente.

- Como regra geral: nunca utilize `from alguma_biblioteca import *` nos programas que serão usados posteriormente. Isso pode funcionar no prompt de comando, entretanto, não se deve usar, pois pode representar uma série de potenciais bugs em seu código.

Nos exemplos a seguir, geralmente usamos a interface `pylab` para as rotinas para fazer gráficos, mas isso é puramente uma questão de gosto e hábito e de nenhuma maneira a única maneira (note que os autores da Matplotlib recomendam o estilo de importação como no exemplo 1a, veja Também [Matplot FAQ entrada](http://matplotlib.org/faq/): [Matplotlib, pylab, e pyplot: como eles estão relacionados?](http://matplotlib.org/faq/usage_faq.html#matplotlib-pyplot-and-pylab-how-are-they-related))

### Salvando a figura em um arquivo

Depois de ter criado a figura (usando o comando `plot`) e adicionado quaisquer rótulos, lengendas etc, você tem duas opções para salvar o gráfico.

1. Você pode exibir a figura (usando `show`) e *interativamente* salvá-lo clicando no ícone do disco.

2. Você pode (sem exibir a figura) salvá-lo diretamente do seu código Python. O comando a ser usado é o `savefig`. O formato é determinado pela extensão do nome do arquivo fornecido. Aqui está um exemplo (`pylabsavefig.py`) que salva o gráfico mostrado na Figura \[fig:pylab1\]


````python
import pylab
import numpy as np

x0 =-np.pi
xf = np.pi
n = 1001
x = np.linspace(x0,xf,n)
#x = np.arange(-3.14, 3.14, 0.01)
y = np.sin(x)

pylab.plot(x, y, label='sin(x)')
pylab.savefig('grafico-seno.png')  # salva o arquivo no formato png
pylab.savefig('grafico-seno.eps')  # salva o arquivo no formato eps
pylab.savefig('grafico-seno.pdf')  # salva o arquivo no formato pdf
pylab.savefig('grafico-seno.svg')  # salva o arquivo no formato svg
pylab.show()
pylab.close()
````

Uma nota sobre os formatos de arquivo: Escolha o formato de arquivo `png` se você planeja incluir seu gráfico em um documento do tipo suite de escritório ou em uma página da Web. Escolha o formato de arquivo `eps` ou` pdf` se você pretende incluir a figura em um documento de Latex - dependendo se você deseja compilá-lo usando `latex` (precisa `eps`) ou pdflatex (pode usar `pdf` \[Melhor\] ou `png`). Se a versão da sua suite de escritório (ou outro software de processamento de texto que você usa) puder manipular arquivos `pdf`, é melhor usar `pdf` do que `png`.

Tanto `pdf` como` eps` são formatos de arquivo vetorial, o que significa que é possível ampliar a imagem sem perder a qualidade (as linhas ainda serão nítidas). Os formatos de arquivo como `png` (e `jpg`, `gif`,` tif`, `bmp`) salva a imagem na forma de um bitmap (ou seja, uma matriz de valores de cor) e aparecerá embaçado ou pixelizado ao ampliar (ou quando impresso em alta resolução).


### Modo interativo

A Matplotlib pode ser executada em dois modos:

- não-interativo (este é o padrão)

- interativo.

No modo não interativo, nenhum gráfico será exibido até que o comando `show()` tenha sido emitido. Neste modo, o comando `show()` deve ser a última instrução do seu programa.

No modo interativo, as parcelas serão imediatamente exibidas após o comando `plot` ter sido emitido.

É possível alternar o modo interativo usando `pylab.ion()` e desativar usando `pylab.ioff()`. A mágica `%matplotlib` do IPython também permite o modo interativo.


### Ajustes finos dos  gráficos 

A Matplotlib nos permite ajustar nossos gráficos em inúmeros detalhes. Aqui está um exemplo:

````python
import pylab
import numpy as np


x0 = -3.14     # Valor inicial de x
xf = +3.14     # Valor final de x
dx = 0.01      # Passo da discretização dx = (xf - x0)/(np-1)

x = np.arange(x0, xf, dx)           # cria um vetor de dados x-data

y1 = np.sin(x)
y2 = np.cos(x)

#Define-se o intervalo de cada eixo
xmin = -3.5   
xmax = +3.5
ymin = -1.1
ymax = +1.1
v = [xmin, xmax, ymin, ymax]                # Vetor que define a escala do gráfico

# Define o tamanho do gráfico
largura = 10  # Largura da figura
altura  =  6  # Altura da figura
pylab.figure(figsize =(largura, altura))     # Define o tamanho do gráfico
# Faz o gráfico das funções
pylab.plot(x, y1, label='sen(x)')
pylab.plot(x, y2, label='cos(x)')
# Altera propriedades do gráfico
pylab.axis(v)                      # Define a escalas x e y do gráfico
pylab.legend()                     # Habilita as legendas
pylab.grid(b='on',axis='both')     # Habilita a grade 
#pylab.grid()                      # Habilita a grade 
pylab.xlabel('x')
pylab.ylabel('y(x)')

pylab.title('Funções seno e cosseno')
````

````python
pylab.figure?
````

Mostrando alguns outros comandos úteis:

- `figure(figsize =(largura, altura))` define o tamanho de figura para largura x altura em polegadas

- `plot(x, y1, label='sen(x)')` A palavra-chave "label" define o nome da curva que será traçada. O rótulo da linha será mostrado na legenda se o comando `legend()` for usado posteriormente.

- Note que chamar o comando `plot` repetidamente, permite que você sobreponha um número qualquer de curvas.

- `axis([x0, xf, y0, yf])` Isso fixa a área exibida para ir de xmin = x0 a xmax = xf na direção x e de ymin = y0 para ymax = yf na direção y.

- `legend()` Este comando exibirá uma legenda com os rótulos, conforme definido no comando plot. Experimente `help (pylab.legend)` para saber mais sobre o posicionamento da legenda.

- `grid()` Este comando exibirá uma grade no pano de fundo.

- `xlabel ('...')` e `ylabel ('...')` permitem rotular os eixos.

Observe além do que você pode escolher diferentes estilos de linha, espessuras de linha, símbolos e cores para os dados a serem plotados. (A sintaxe é muito semelhante ao MATLAB.) Por exemplo:

- `plot(x, y, 'og')` desenhará círculos (`o`) em verde (`g`)

- `plot (x, y, '- r')` traçará uma linha (`-`) em vermelho (`r`)

- `plot (x, y, '- b', linewidth = 2)` traçará uma linha (`-`) azul (`b`) com espessura de dois pixels `linewidth=2` que é duas vezes maior do que o padrão.

### Fazendo o gráfico de mais de uma curva

#### Duas (ou mais) curvas em um gráfico

Chamando o comando `plot` repetidamente, mais de uma curva pode ser desenhada no mesmo gráfico. Exemplo:

## Legendas e LaTeX

````python
"""
Gráfico demonstrando a integral como a área sob uma curva.

Embora este seja um exemplo simples, ele demonstra alguns ajustes importantes:

     * Um gráfico de linha simples com cor personalizada e largura de linha.
     * Uma região sombreada criada usando um Polygon patch.
     * Uma etiqueta de texto com renderização mathtext.
     * chamada figtext para rotular os eixos x e y.
     * Uso dos spines do eixo para ocultar as espinhas superior e direita.
     * Personalizando a colocação dos ticks e rótulos.
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon


def func(x):
    return (x - 3) * (x - 5) * (x - 7) + 85


a, b = 2, 9  # Limites da integral
x = np.linspace(0, 10)
y = func(x)


largura = 10  # Largura da figura
altura  = 6  # Altura da figura
#
fig, ax = plt.subplots(figsize =(largura, altura))
#
plt.plot(x, y, 'r', linewidth=2)
plt.ylim(ymin=0)

# Tornar a região sombreada
ix = np.linspace(a, b)
iy = func(ix)
verts = [(a, 0)] + list(zip(ix, iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a + b), 30, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.05, '$x$')
plt.figtext(0.1, 0.9, '$y$')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([])

plt.show()
````

## Dipolo elétrico

````python
"""
Demonstra o cáculo do gradiente com a matplotlib.tri.CubicTriInterpolator.
"""
from matplotlib.tri import Triangulation, UniformTriRefiner,\
    CubicTriInterpolator
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import math


#-----------------------------------------------------------------------------
# Potencial elétrico de um dipolo
#-----------------------------------------------------------------------------
def dipole_potential(x, y):
    """ The electric dipole potential V """
    r_sq = x**2 + y**2
    theta = np.arctan2(y, x)
    z = np.cos(theta)/r_sq
    return (np.max(z) - z) / (np.max(z) - np.min(z))


#-----------------------------------------------------------------------------
# Criando uma triangulação
#-----------------------------------------------------------------------------
# Primeiro crie as coordenadas x e y dos pontos.
n_angles = 30
n_radii = 10
min_radius = 0.2
radii = np.linspace(min_radius, 0.95, n_radii)

angles = np.linspace(0, 2*math.pi, n_angles, endpoint=False)
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)
angles[:, 1::2] += math.pi/n_angles

x = (radii*np.cos(angles)).flatten()
y = (radii*np.sin(angles)).flatten()
V = dipole_potential(x, y)

# Criar a Triangulação; Sem triângulos especificados para a 
# triangulação Delaunay criado
triang = Triangulation(x, y)

# Mascara fora de triângulos indesejados.
xmid = x[triang.triangles].mean(axis=1)
ymid = y[triang.triangles].mean(axis=1)
mask = np.where(xmid*xmid + ymid*ymid < min_radius*min_radius, 1, 0)
triang.set_mask(mask)

#-----------------------------------------------------------------------------
# Refinar dados - interpola o potencial elétrico V
#-----------------------------------------------------------------------------
refiner = UniformTriRefiner(triang)
tri_refi, z_test_refi = refiner.refine_field(V, subdiv=3)

#-----------------------------------------------------------------------------
# Calcula o campo elétrico (Ex, Ey) como gradiente de potencial elétrico
#-----------------------------------------------------------------------------
tci = CubicTriInterpolator(triang, -V)
# Gradiente solicitado aqui nos nós do mesh, mas poderia ser em qualquer outro lugar:
(Ex, Ey) = tci.gradient(triang.x, triang.y)
E_norm = np.sqrt(Ex**2 + Ey**2)

#-----------------------------------------------------------------------------
# Faz o gráfico da triangulação, o potencial iso-contornos e o campo vetorial
#-----------------------------------------------------------------------------

largura = 10  # Largura da figura
altura  = 10  # Altura da figura
#plt.figure(figsize =(largura, altura))     # Define o tamanho do gráfico

fig, ax = plt.subplots(figsize =(largura, altura))
ax.set_aspect('equal')
# Aplicar as margens e ampliá elas para dar espaço aos vetores.
ax.use_sticky_edges = False
ax.margins(0.07)

ax.triplot(triang, color='0.8')

levels = np.arange(0., 1., 0.01)
cmap = cm.get_cmap(name='hot', lut=None)
ax.tricontour(tri_refi, z_test_refi, levels=levels, cmap=cmap,
              linewidths=[2.0, 1.0, 1.0, 1.0])
# Faz um gráfico da direção do vetor campo elétrico 
ax.quiver(triang.x, triang.y, Ex/E_norm, Ey/E_norm,
          units='xy', scale=10., zorder=3, color='blue',
          width=0.007, headwidth=3., headlength=4.)

ax.set_title('Gráfico do gradiente: um dipolo elétrico')


plt.show()
````

## Função integral do cosseno e seno

Esse exemplo de [integrais trigonométicas](https://en.wikipedia.org/wiki/Trigonometric_integral). Essa função é definida por

$$\operatorname{Ci}(x) = -\int_x^\infty \frac{\cos t}{t}\,dt = \gamma + \ln x + \int_0^x \frac{\cos t - 1}{t}\,dt
\qquad (|{\rm Arg}(x)|<\pi)$$


Consulte a biblioteca da sipy para [funções especiais](https://docs.scipy.org/doc/scipy/reference/special.html)


[integral do cosseno](https://commons.wikimedia.org/wiki/File:Cosine_integral.svg)


A localização da legenda pode ser especificada definindo-se loc com uma tupla de 2 floats, que é interpretada como o canto inferior esquerdo da legenda na coordenada de eixos normalizados.

> codes = {'best': 0, 'upper right': 1, 'upper left': 2, 'lower left': 3, 'lower right': 4, 'right': 5, 'center left': 6, 'center right': 7, 'lower center': 8, 'upper center': 9, 'center': 10}

````python
%matplotlib inline
from numpy import *
from scipy.special import sici
from matplotlib.pyplot import *

# Calculando o valor das funções
x = linspace(0.001, 25, 100)   # definino o vetor x
(y_sine,y_cosine) = sici(x);   # Calcula o valor das funções SI(x) e CI(x)


# Defini a Matriz Nlinhas x Ncolunas que será mostrado os gráficos
Nlinhas, Ncolunas = 1, 2  

largura = 14  # Largura da figura
altura  = 6  # Altura da figura

fig = figure(figsize=(largura, altura)) # Define o tamanho do gráfico


# Integral do seno
ax = fig.add_subplot(Nlinhas,Ncolunas,1) # Primeiro gráfico da Matriz Nlinhas x Ncolunas
ax.set_ylim(0, 2)
# Aqui fazemos uso do LaTeX para expressar a legenda da figura
ax.plot(x,y_sine, linewidth=2, label=r"Si$(x)=\int_0^x\frac{\sin(t)}{t}dt$")
ax.plot([0, 25], [1.58, 1.58], linestyle='--', color='g', label=r"asymptote");
ax.set_ylabel(r"Si($x$)")
ax.set_xlabel(r"$x$")
ax.grid(True)
ax.minorticks_on()
legend(loc=4)
fig.savefig("Sine integral.svg",bbox_inches="tight",\
        pad_inches=.15)


# Integral do cosseno
ax = fig.add_subplot(Nlinhas,Ncolunas,2) # Segundo gráfico da Matriz Nlinhas x Ncolunas
ax.set_ylim(-6.5, 0.75)
# Aqui fazemos uso do LaTeX para expressar a legenda da figura
ax.plot(x,y_cosine, linewidth=2, label=r"Ci$(x)=\gamma+\ln(x)+\int_0^x\frac{\cos(t)-1}{t}dt$")
ax.plot([0, 25], [0, 0], linestyle='--', color='g', label=r"asymptote");
ax.set_ylabel(r"Ci($x$)")
ax.set_xlabel(r"$x$")
ax.grid(True)
ax.minorticks_on()
fig.savefig("Cosine integral.svg",bbox_inches="tight",\
        pad_inches=.15)
legend(loc=4)
show()
````

## Curvas Preenchidas

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2*np.pi*x)
y2 = 1.2*np.sin(4*np.pi*x)

largura = 10  # Largura da figura
altura  = 10  # Altura da figura

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True,figsize =(largura, altura))
# Controlando a distância entre os eixos
plt.subplots_adjust(hspace=0.001)

ax1.fill_between(x, 0, y1)
ax1.set_ylabel('between y1 and 0')

ax2.fill_between(x, y1, 1)
ax2.set_ylabel('between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_ylabel('between y1 and y2')
ax3.set_xlabel('x')

# Agora preencha a região entre y1 e y2 a qual uma condição lógica é satisfeita. 
# Observe que isso é diferente de chamar "fill_between(x [where], y1 [where], y2 [where]" 
# devido aos efeitos de borda em várias regiões contínuas.

fig, (ax, ax1) = plt.subplots(2, 1, sharex=True, figsize =(largura, altura))
ax.plot(x, y1, x, y2, color='black')
ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax.set_title('fill between where')

# Suporte de teste para matrizes mascaradas.
y2 = np.ma.masked_greater(y2, 1.0)
ax1.plot(x, y1, x, y2, color='black')
ax1.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
ax1.fill_between(x, y1, y2, where=y2 <= y1, facecolor='red', interpolate=True)
ax1.set_title('Now regions with y2>1 are masked')

# Este exemplo ilustra um problema; Devido à grade de dados, existem triângulos 
# não preenchidos indesejados nos pontos de cruzamento. Uma solução de força bruta 
# seria interpolar todas as matrizes para uma grade muito fina antes de plotar.

# Mostram como usar transformações para criar eixos de spans onde uma determinada 
# condição é satisfeita

largura = 10  # Largura da figura
altura  = 6  # Altura da figura

fig, ax = plt.subplots(figsize =(largura, altura))
y = np.sin(4*np.pi*x)
ax.plot(x, y, color='black')

# Use as coordenadas de dados para o eixo x e as coordenadas dos eixos para o eixo y
import matplotlib.transforms as mtransforms
trans = mtransforms.blended_transform_factory(ax.transData, ax.transAxes)
theta = 0.9
ax.axhline(theta, color='green', lw=2, alpha=0.5)
ax.axhline(-theta, color='red', lw=2, alpha=0.5)
ax.fill_between(x, 0, 1, where=y >  theta, facecolor='green', alpha=0.5, transform=trans)
ax.fill_between(x, 0, 1, where=y < -theta, facecolor='red',   alpha=0.5, transform=trans)


plt.show()
````

## Linhas multicoloridas

````python
'''
Partes de uma linha colorida com base nas suas propriedades, por exemplo, inclinação.
'''

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap, BoundaryNorm

x = np.linspace(0, 3 * np.pi, 500)
y = np.sin(x)
z = np.cos(0.5 * (x[:-1] + x[1:]))  # first derivative

# Cria um mapa de cores para vermelho, verde e azul e uma norma para colorir
# f' < -0.5 red, f' > 0.5 blue, e o resto green
cmap = ListedColormap(['r', 'g', 'b'])
norm = BoundaryNorm([-1, -0.5, 0.5, 1], cmap.N)

# Criar um conjunto de segmentos de linha para que possamos colori-los individualmente 
# Isso cria os pontos como uma matriz N x 1 x 2 para que possamos empilhar pontos 
# facilmente para obter os segmentos. A matriz de segmentos para a coleção de linhas 
# precisa ser numlines x point por linha x 2 (x e y)
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

# Crie o objeto de coleção de linha, definindo os parâmetros de mapeamento de cores.
# Tem de definir os valores reais utilizados para o mapeamento de cores separadamente.
lc = LineCollection(segments, cmap=cmap, norm=norm)
lc.set_array(z)
lc.set_linewidth(3)

# Tamanho das figuras
largura = 10  # Largura da figura
altura  = 6  # Altura da figura

fig1 = plt.figure(figsize =(largura, altura))
plt.gca().add_collection(lc)
plt.xlim(x.min(), x.max())
plt.ylim(-1.1, 1.1)

# Agora, faça um segundo gráfico colorindo a curva usando um mapa de cores contínuo
t = np.linspace(0, 10, 200)
x = np.cos(np.pi * t)
y = np.sin(t)
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)

lc = LineCollection(segments, cmap=plt.get_cmap('copper'),
                    norm=plt.Normalize(0, 10))
lc.set_array(t)
lc.set_linewidth(3)

fig2 = plt.figure(figsize =(largura, altura))
plt.gca().add_collection(lc)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()
````

## Equação de Van der Pol


Esse exemplo, foi extraído da página [Equação de Van der Pol](https://pt.wikipedia.org/wiki/Equa%C3%A7%C3%A3o_de_Van_der_Pol). 

Esse é um exemplo de solução numérica do oscilador de Van der Pol, com $\mu = 5$.  O intervalo de tempo entre cada círculo é o mesmo e suas cores são proporcionais a velocidade do oscilador no instante de tempo que representam. Círculos vermelhos possuem velocidades maiores que 0.25 vezes a velocidade máxima, círculos azuis possuem velocidades menores que 0.25 vezes a velocidade minima e círculos verdes possuem velocidades entre esses valores.
A '''equação de Van der Pol''' representa um [Vibração|oscilador] com um termo de amortecimento [Sistema dinâmico não linear|não linear]. Sua evolução no tempo é descrita pela [equação diferencial] de segunda ordem:

$$\frac{dx^2}{dt} - \mu(1-x^2)\frac{dx}{dt} + x = 0$$

onde $x(t)$  é a posição do oscilador e $\mu$ mede a força do termo de amortecimento. No caso especial $\mu = 0$ recuperamos um [oscilador harmônico] simples.

### História

O oscilador de Van der Pol foi proposto em 1920 por [Balthasar van der Pol].


````python
%matplotlib inline
from numpy import *
from scipy import *
from scipy.integrate import odeint
from matplotlib.pyplot import *
from mpl_toolkits.axes_grid.axislines import SubplotZero

def myFun(u,t=0.,mu=.5):
    x = u[0]
    v = u[1]
    dx = v
    dv = mu*(1.-x**2)*v-x
    return (dx,dv)

t = linspace(-5.5,5.5,300)
u0 = array([1.,1.2])
mu = [.1, .2, .5, 1., 1.5, 2.0, 2.5]
fig = figure(figsize=(10,7))
ax = SubplotZero(fig,111)
fig.add_subplot(ax)
ax.grid(True)
for direction in ["right","top"]:
        ax.axis[direction].set_visible(False)
for m in mu:
    u = odeint(myFun,u0,t,args=(m,))
    ax.plot(u[:,0],u[:,1],lw=1.5,label=r'$\mu=%.1f$'%m)
x = linspace(-3,3,15)
y = linspace(-4,4,15)
x,y = meshgrid(x,y)
X,Y = myFun([x,y])
M = (hypot(X,Y))
M[M==0]=1.
X,Y = X/M, Y/M
ax.quiver(x,y,X,Y,M,pivot='mid',cmap=cm.jet)
ax.minorticks_on()
ax.legend(handletextpad=0,loc='upper left')
setp(ax.get_legend().get_texts(),fontsize=12)
fig.savefig("Van_der_pols_equation_phase_portrait.svg",bbox_inches="tight",\
        pad_inches=.15)
show()
````

````python
import numpy as np
import matplotlib.mlab as mlab    # Comando de compatibilidade com o Matlab
import matplotlib.pyplot as plt
import matplotlib.cm as cm        # Submodulo do mapa de cores

#Cria uma matriz Z que contém alguns dados interessantes
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y  = np.meshgrid(x, y)
Z = mlab.bivariate_normal(X, Y, 3.0, 1.0, 0.0, 0.0)


fig = plt.figure(figsize=(10,8))

Nx, Ny = 2, 3
plt.subplot(Nx, Ny, 1)  # o próximo plot será mostrado no
                        # primeiro subplot da matriz Nx x Ny
                        # dos subplots
        
plt.imshow(Z, cmap=cm.jet)  # padrão: colourmap 'jet'
plt.title("Mapa de cores  jet")

plt.subplot(Nx, Ny, 2)   # o próximo plot para o segundo subplot
plt.imshow(Z, cmap=cm.jet_r)  # cores reversa as do jet
plt.title("Mapa de cores  jet_r")

plt.subplot(Nx, Ny, 3)
plt.imshow(Z, cmap=cm.gray)
plt.title("Mapa de cores  gray")

plt.subplot(Nx, Ny, 4)
plt.imshow(Z, cmap=cm.hsv)
plt.title("Mapa de cores  hsv")

plt.subplot(Nx, Ny, 5)
plt.imshow(Z, cmap=cm.gist_earth)
plt.title("Mapa de cores  gist_earth")

plt.subplot(Nx, Ny, 6)
# Faz isolines reduzindo o número de cores para 10
mycmap = cm.get_cmap('jet', 10)    # 10 cores discretas
plt.imshow(Z, cmap=mycmap)
plt.title(u"Mapa de cores jet\n(só 10 cores)")

plt.savefig("pylab-imshowcm.pdf")
````

````python
# linha3d.py
# -------------------------------------------------------------------------
# Cria um gráfico paramétrico tridimensional 
# ------------------------------------------------------------------------- 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D     # importa as bibliotecas 3D 

fig = plt.figure()                  # cria uma nova figura
ax = Axes3D(fig)                    # cria um objeto gráfico 3D anexado a figura
t = np.linspace(0, 5*np.pi, 501)    # define os parametros para o gráfico paramétrico 3D
x = t * np.cos(t)
y = t * np.sin(t)

ax.plot(x, y, t)        # gera o gráfico 3D
plt.show()
````

Note que foi usado o método `plot` para anexar a um objeto específico. Em seguida criou-se o objeto `Axes3D`, o que  permitiu usar os seus métodos para gerar gráficos tridimensionais. O método `ax.plot` aceita três arrays, cujas entradas correspondentes são interpretadas como as coordenadas $x$, $y$ e $z$ dos pontos a serem desenhados. O objeto gerado pelo `Axes3D (fig)` também pode criar gráficos de superfície, gráficos de contornos tridimensionais, histogramas tridimensionais e muito mais.

Sua tela é bidimensional. Se quer um gráfico "3D", o que você realmente obtém deve ser uma projeção bidimensional - o que uma câmera veria em sua malha de algum ponto de vista externo. Alternar para um ponto de vista diferente pode tornar mais fácil ver o que está acontecendo. Usando o magic `%matplotlib qt` no Jupyter/IPython ele fara o gráfico em uma janela aparte na qual pode-se clicar e arrastar dentro de seu gráfico para alterar o ponto de vista. (Isso depende da parte traseira dos gráficos que você está usando.) Se isso não for possível, você pode alterar o ponto de vista do prompt de comando do Jupyter/IPython ou em um script com o comando `ax.view_init(elev = 30, azim = 30)`. Você especifica o ponto de vista dando a elevação e azimute, em graus - não radianos.

````python
# Gráfico do Atrator de Lorenz baseado no artigo de Edward Lorenz de 1963 "Deterministic
# Nonperiodic Flow".
# http://journals.ametsoc.org/doi/abs/10.1175/1520-0469%281963%29020%3C0130%3ADNF%3E2.0.CO%3B2
#
# Nota: Delvido a ela ser uma simples EDO não-linear, seria mais fácil implementar
#       usando a SciPy ode solver, mas a abordagem atual depende somente da Numpy.
 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def lorenz(x, y, z, s=10, r=28, b=2.667):
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01
stepCnt = 10000

# Precisa de mais um para os valores iniciais
xs = np.empty((stepCnt + 1,))
ys = np.empty((stepCnt + 1,))
zs = np.empty((stepCnt + 1,))

# Definição de valores iniciais
xs[0], ys[0], zs[0] = (0., 1., 1.05)

# Passando pelo "tempo".
for i in range(stepCnt):
    # Derivadas dos estados de X, Y, Z
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)

fig = plt.figure(figsize=(8,6))
ax = fig.gca(projection='3d')

ax.plot(xs, ys, zs, lw=0.5)
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")
ax.set_zlabel("Eixo Z")
ax.set_title(" Atrator de Lorenz")

plt.show()
````

````python
%matplotlib inline
from __future__ import division
import numpy as np
from numpy import pi, cos, sin, sqrt, outer, ones
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
u = np.linspace(0, 2 * pi, 100)
v = np.linspace(0, pi, 100)
one_v = ones(100)


# Esfera
x1 = 10 * outer(cos(u), sin(v))
y1 = 10 * outer(sin(u), sin(v)) + 17
z1 = 10 * outer(one_v, np.cos(v))
ax.plot_surface(x1, y1, z1, rstride=5, cstride=5, cmap='Spectral',
                linewidth=0.5)

# Cilindro
x2 = 6 * outer(cos(u), one_v)
y2 = 6 * outer(sin(u), one_v)
z2 = 20 * outer(one_v, v/pi - 0.5)
ax.plot_surface(x2, y2, z2, rstride=5, cstride=5, cmap='Spectral',
                linewidth=0.5)

# Hiperbolóide
v = 2*v/pi - 1
x3 = 6 * outer(cos(u), sqrt(1 + v**2))
y3 = 6 * outer(sin(u), sqrt(1 + v**2)) - 16
z3 = 12 * outer(one_v, v)
ax.plot_surface(x3, y3, z3, rstride=5, cstride=5, cmap='Spectral',
                linewidth=0.5)

# Fixa a razão entre tamanhos e alguns detalhes dos eixos
ax.set_xlim(-13, 13)
ax.set_ylim(-13, 13)
ax.set_zlim(-13, 13)
ax.view_init(elev=35, azim=-45)
plt.axis('off')
plt.savefig('Gaussian_curvature.svg', transparent=True)
plt.show()
````

### Animação

A maneira mais fácil de fazer uma animação na matplotlib é usar uma das classes de animação.

- `FuncAnimation` Faz uma animação chamando repetidamente uma função func.

- `ArtistAnimation` Anima usando um conjunto fixo de objetos `Artist`.

Em ambos os casos, é fundamental manter uma referência ao objeto de instância. A animação é avançada por um temporizador (normalmente a partir da estrutura de GUI do ambiente usado) na qual o objecto de `Animation` contém a referência apenas para ele. Se você não mantiver uma referência ao objeto `Animation`, ele (e, portanto, os temporizadores), será coletado lixo o que irá parar a animação.

### Para salvar uma animação no disco

- `Animation.save` Salva um arquivo de filme desenhando cada quadro.

- `Animation.to_html5_video` Retorna a animação como uma tag de vídeo HTML5.

Consulte [Classes auxiliares](http://matplotlib.org/api/animation_api.html#ani-writer-classes) abaixo para obter detalhes sobre quais formatos de filme são suportados.

### FuncAnimation

O funcionamento interno da FuncAnimation é mais ou menos:

```python
for d in frames:
   artists = func(d, *fargs)
   fig.canvas.draw_idle()
   plt.pause(interval)
```

Com detalhes para lidar com "blitting" (para melhorar drasticamente o desempenho ao vivo), para ser não-bloqueado, para lidar com repetições, com vários eixos animados e salvar facilmente a animação para um arquivo de filme.

"Blitting" é uma técnica antiga em computação gráfica. A essência geral é tomar um mapa de bits existente (no nosso caso uma figura principalmente rasterizada) e, em seguida, 'blit' mais um artista no topo. Assim, ao gerenciar um bitmap "limpo" salvo, só podemos re-desenhar os poucos artistas que estão mudando em cada quadro e, possivelmente, economizar quantidades significativas de tempo. Ao usar blitting (passando blit = True) o núcleo do loop de FuncAnimation fica um pouco mais complicado

```python
ax = fig.gca()

def update_blit(artists):
    fig.canvas.restore_region(bg_cache)
    for a in artists:
        a.axes.draw_artist(a)

    ax.figure.canvas.blit(ax.bbox)

artists = init_func()

for a in artists:
   a.set_animated(True)

fig.canvas.draw()
bg_cache = fig.canvas.copy_from_bbox(ax.bbox)

for f in frames:
    artists = func(f, *fargs)
    update_blit(artists)
    plt.pause(interval)
```

Isto é, naturalmente, deixar de fora muitos detalhes (como a atualização do plano de fundo quando a figura é redimensionada ou totalmente re-desenhado). No entanto, neste exemplo minimalista  espero dár uma idéia de como as funções `init_func` e `func` são usados dentro de `FuncAnimation` e um pouco da teoria de como 'blitting' funciona.

A assinatura esperada em `func` e `init_func` é muito simples para manter `FuncAnimation` fora de sua lógica de consertos e tramas, mas isso significa que os objetos que você pode passar devem saber em que artistas eles devem estar trabalhando. Existem várias abordagens para lidar com isso, de complexidade variável e encapsulamento. A abordagem mais simples, que funciona muito bem no caso de um script, é definir o artista em um escopo global e deixar Python resolver as coisas. 

Para incluir uma animação no notebook, siga, as instruções abaixo, maiores detalhes podem ser encontrados no blog [Embedding Matplotlib Animations in Jupyter Notebooks](http://louistiao.me/posts/notebooks/embedding-matplotlib-animations-in-jupyter-notebooks/) do Louis Tiao.

````python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
````

````python
# Primeiro configure a figura, o eixo e o elemento de gráfico que queremos animar
fig, ax = plt.subplots(figsize=(10,6))

ax.set_xlim(( 0, 2))
ax.set_ylim((-1.1, 1.1))

line, = ax.plot([], [], lw=2)
````

````python
# Função de inicialização: plot o plano de fundo de cada quadro
def init():
    line.set_data([], [])
    return (line,)
````

````python
# Função animação. Essa é chamada sequencialmente
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
````

````python
# Chama o animador. blit=True significa apenas re-desenhar as peças que foram alteradas.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=40, blit=True, repeat=False)

# Em um programa python fora do ambiente do notebook
# chame o vídeo:
# anim
````

````python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import animation, rc
from IPython.display import HTML

u=np.linspace(0, 2*np.pi, 128)
v=np.sin(u)

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro', animated=True)


def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1.05, 1.05)
    ln, = plt.plot(u, v, 'g', animated=True)
    return ln,

def update(frame):
    xdata.append(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln,

# para gerar a animação no notebook
rc('animation', html='html5')

ani = FuncAnimation(fig, update, frames=np.linspace(0, 2*np.pi, 80),
                    init_func=init, blit=True, repeat=False)

ani

#plt.show()
````

````python

""" Faz uma animação dos passo de um passeio aleatório bidimensional. """
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from numpy.random import random as rand

# Define o número de passos para cada passeio aleatório.
n_passos = 150

# Cria uma figura vazia com o tamanho desejado.
limite = 20
fig = plt.figure()      # must have figure object for movie
ax = plt.axes(xlim=(-limite, limite), ylim=(-limite, limite))

# Cria uma linha e um ponto sem dados.  Eles serão adaptados
# durante cada frame da animação. 
(a_linha,) = ax.plot([], [], lw=2)             # linha para mostrar a trajetória
(o_ponto,) = ax.plot([], [], 'ro', ms=9)       # ponto para mostra a posição atual

# Gera os dados do passeio aleatório.
x_passos = 2*(rand(n_passos) < 0.5) - 1     # gera passos aleatórios +/- 1
y_passos = 2*(rand(n_passos) < 0.5) - 1
x_coordenada = x_passos.cumsum()             # soma os passos para obter a posição
y_coordenada = y_passos.cumsum()

# Essa função irá gerar cada frame da animação.
# Ela adiciona todos os dados através do frame n para uma linha
# e move um ponto para a n-ésima posição do passeio. 
def get_step(n, x, y, Essa_linha, Esse_ponto):
    Essa_linha.set_data(x[:n+1], y[:n+1])
    Esse_ponto.set_data(x[n], y[n])

# Chama a função animation.FuncAnimation e cria o filme.
anim = animation.FuncAnimation(fig, get_step, frames=n_passos, \
                    fargs=(x_coordenada,y_coordenada,a_linha,o_ponto) )

# Salva o filme no diretório atual.
# *** ESTA CAUSARÁ UM ERRO A MENOS QUE FFMPEG OU MENCODER ESTEJAM INSTALADOS. ***
# my_movie.save('random_walk.mp4', fps=30)

HTML(anim.to_html5_video())
````

O segundo método é para nós functools.partial para 'ligar' artistas para funcionar. Um terceiro método é usar fechamentos para construir os artistas necessários e funções. Um quarto método é criar uma classe.

````python
"""
=====
Decaimento
=====

Esse exemplo mostra uma animação de um decaimento sinusoidal.
"""


import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


def data_gen(t=0):
    cnt = 0
    while cnt < 1000:
        cnt += 1
        t += 0.1
        yield t, np.sin(2*np.pi*t) * np.exp(-t/10.)


def init():
    ax.set_ylim(-1.1, 1.1)
    ax.set_xlim(0, 10)
    del xdata[:]
    del ydata[:]
    line.set_data(xdata, ydata)
    return line,

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [], []


def run(data):
    # update the data
    t, y = data
    xdata.append(t)
    ydata.append(y)
    xmin, xmax = ax.get_xlim()

    if t >= xmax:
        ax.set_xlim(xmin, 2*xmax)
        ax.figure.canvas.draw()
    line.set_data(xdata, ydata)

    return line,

ani = animation.FuncAnimation(fig, run, data_gen, blit=True, interval=40,
                              repeat=False, init_func=init)

ani
#plt.show()
````

````python
"""
===========================
O problema do pendulo duplo
===========================

Essa animação ilustra o problema do pendulo duplo.
"""

# Pendulo duplo, o foi código foi adaptado do seguinte código em C:
# http://www.physics.usyd.edu.au/~wheat/dpend_html/solve_dpend.c

from numpy import sin, cos
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
import matplotlib.animation as animation

from matplotlib import animation, rc
from IPython.display import HTML

G  = 9.8  # acceleration due to gravity, in m/s^2
L1 = 1.0  # length of pendulum 1 in m
L2 = 1.0  # length of pendulum 2 in m
M1 = 1.0  # mass of pendulum 1 in kg
M2 = 1.0  # mass of pendulum 2 in kg


def derivs(state, t):

    dydx = np.zeros_like(state)
    dydx[0] = state[1]

    del_ = state[2] - state[0]
    den1 = (M1 + M2)*L1 - M2*L1*cos(del_)*cos(del_)
    dydx[1] = (M2*L1*state[1]*state[1]*sin(del_)*cos(del_) +
               M2*G*sin(state[2])*cos(del_) +
               M2*L2*state[3]*state[3]*sin(del_) -
               (M1 + M2)*G*sin(state[0]))/den1

    dydx[2] = state[3]

    den2 = (L2/L1)*den1
    dydx[3] = (-M2*L2*state[3]*state[3]*sin(del_)*cos(del_) +
               (M1 + M2)*G*sin(state[0])*cos(del_) -
               (M1 + M2)*L1*state[1]*state[1]*sin(del_) -
               (M1 + M2)*G*sin(state[2]))/den2

    return dydx

# create a time array from 0..100 sampled at 0.05 second steps
dt = 0.05
t = np.arange(0.0, 50, dt)

# th1 and th2 are the initial angles (degrees)
# w10 and w20 are the initial angular velocities (degrees per second)
th1 = 120.0
w1  = 0.0
th2 = -10.0
w2  = 0.0

# initial state
state = np.radians([th1, w1, th2, w2])

# integrate your ODE using scipy.integrate.
y = integrate.odeint(derivs, state, t)

x1 = L1*sin(y[:, 0])
y1 = -L1*cos(y[:, 0])

x2 = L2*sin(y[:, 2]) + x1
y2 = -L2*cos(y[:, 2]) + y1

fig = plt.figure()
ax = fig.add_subplot(111, autoscale_on=False, xlim=(-2, 2), ylim=(-2, 2))
ax.grid()

line, = ax.plot([], [], 'o-', lw=2)
time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)


def init():
    line.set_data([], [])
    time_text.set_text('')
    return line, time_text


def animate(i):
    thisx = [0, x1[i], x2[i]]
    thisy = [0, y1[i], y2[i]]

    line.set_data(thisx, thisy)
    time_text.set_text(time_template % (i*dt))
    return line, time_text

# para gerar a animação no notebook
rc('animation', html='html5')

pendulo_duplo_animado = animation.FuncAnimation(fig, animate, np.arange(1, len(y)),
                              interval=25, blit=True, init_func=init)

pendulo_duplo_animado.save('Pendulo_duplo.mp4', fps=15)
pendulo_duplo_animado
#plt.show()
````
